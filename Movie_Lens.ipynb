```python
# Import necessary libraries
import pandas as pd

# Define file paths
movies_path = 'movies.csv'
ratings_path = 'ratings (3).xlsx'

# Load the movies dataset
movies_df = pd.read_csv(movies_path)

# Load the ratings dataset
ratings_df = pd.read_excel(ratings_path)

# Display the first few rows of each dataset
movies_head = movies_df.head()
ratings_head = ratings_df.head()

movies_head, ratings_head
```

```python
# Perform different types of joins between ratings and movies on 'movieId'

# Inner Join: Keep only matching movieId values
inner_merged = ratings_df.merge(movies_df, on='movieId', how='inner')

# Left Join: Keep all rows from ratings
left_merged = ratings_df.merge(movies_df, on='movieId', how='left')

# Right Join: Keep all rows from movies
right_merged = ratings_df.merge(movies_df, on='movieId', how='right')

# Outer Join: Keep all rows from both datasets
outer_merged = ratings_df.merge(movies_df, on='movieId', how='outer')

inner_merged_head = inner_merged.head()
left_merged_head = left_merged.head()
right_merged_head = right_merged.head()
outer_merged_head = outer_merged.head()

inner_merged_head, left_merged_head, right_merged_head, outer_merged_head
```

```python
import numpy as np

# Define bins for np.digitize
bins = [1, 2, 3, 4, 5]
ratings_df['digitized_rating'] = np.digitize(ratings_df['rating'], bins)

# Define custom bins and labels for pd.cut
cut_bins = [0, 2, 3.5, 4.5, 5]
cut_labels = ['Low', 'Medium', 'High', 'Very High']
ratings_df['cut_rating'] = pd.cut(ratings_df['rating'], bins=cut_bins, labels=cut_labels)

# Use pd.qcut to divide ratings into quantiles
ratings_df['qcut_rating'] = pd.qcut(ratings_df['rating'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Display the transformed ratings DataFrame
ratings_transformed_head = ratings_df.head()

ratings_transformed_head
```
